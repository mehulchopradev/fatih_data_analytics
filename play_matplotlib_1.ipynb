```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
roll = [10, 12, 15, 16]
marks = [67, 100, 56, 80]
```

```python
plt.bar(roll, marks)
plt.ylabel('Marks')
plt.xlabel('Roll')
plt.show()
```
